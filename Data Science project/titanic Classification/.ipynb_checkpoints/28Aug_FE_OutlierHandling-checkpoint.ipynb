# Outlier Detection using Boundary Values



```python
# df['residential_assets_value']
# df['commercial_assets_value']
# df['bank_asset_value']
```

```python
cols = ['residential_assets_value','commercial_assets_value','bank_asset_value']
```

```python
df[cols].describe()
```

```python
for col in cols:
    df[col] = df[col]/10000
```

```python
df[cols].describe()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(df['residential_assets_value'])

plt.figure(figsize=(8,4))
sns.kdeplot(df['commercial_assets_value'])

plt.figure(figsize=(8,4))
sns.kdeplot(df['bank_asset_value'])
```

#### Boundary Values 


(mean + 3*std_dev)

```python
for col in cols:
    print(f'Upper Boundary Value for {col} : {df[col].mean()+3*df[col].std()}')
    print(f'Lower Boundary Value for {col} : {df[col].mean()-3*df[col].std()}')
    print('='*80)
```

```python
# Upper Boundary Value for residential_assets_value 
rav_upper = df['residential_assets_value'].mean()+3*df['residential_assets_value'].std()

# Lower Boundary Value for residential_assets_value 
rav_lower = df['residential_assets_value'].mean()-3*df['residential_assets_value'].std() 
```

```python
print('Upper Val: ',rav_upper)
print('\nLower Val: ',rav_lower)
```

```python
df[(df['residential_assets_value']>rav_upper) | (df['residential_assets_value']<rav_lower)]
```

```python
df[(df['residential_assets_value']>rav_upper) | (df['residential_assets_value']<rav_lower)].shape
```

```python
df.shape
```

# Trimming or Capping:

#### Trimming: Removing the data from the datasets

#### Capping: Replacing the data with the Upper/Lower Value

```python
# Trimming

trimmed_df = df[(df['residential_assets_value']<rav_upper) & (df['residential_assets_value']>rav_lower)]
```

```python
trimmed_df.head()
```

```python
trimmed_df.shape
```

```python
trimmed_df['residential_assets_value'].max()
```

```python
rav_upper
```

```python
rav_lower
```

```python
# Capping

# Upper Boundary Value for residential_assets_value 
rav_upper = df['residential_assets_value'].mean()+3*df['residential_assets_value'].std()

# Lower Boundary Value for residential_assets_value 
rav_lower = df['residential_assets_value'].mean()-3*df['residential_assets_value'].std() 


df['residential_assets_value']  = np.where(df['residential_assets_value']>rav_upper,rav_upper,
                                           np.where(df['residential_assets_value']<rav_lower,rav_lower,
                                                    df['residential_assets_value']))
```

```python
df['residential_assets_value'].describe()
```

# Outlier Detection using Z-score 

### This technique is only used when the data is <u>Normally Distributed</u>

##### But for the sake of learning the technique, we will use this technique on the above columns. 




#### We will only consider values with z-score between -3 and 3 (there maybe situations where this range might change on the context of the data)

```python
df['Z_score_rav'] = (df['residential_assets_value'] - df['residential_assets_value'].mean())/df['residential_assets_value'].std()
```

```python
df[['residential_assets_value','Z_score_rav']].head(10)
```

```python
# Trimming

trimmed_df = df[(df['Z_score_rav']<3) & (df['Z_score_rav']>-3)]

print(trimmed_df.shape)
```
